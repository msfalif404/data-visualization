# Create Multiple Lineplot

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Dataframe

```python
tanggal = pd.date_range(start = '2021-01-01', end = '2024-01-01', freq = 'D')
harga_saham = np.random.randn(len(tanggal)).cumsum() + 100
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'tanggal': tanggal,
    'harga_saham': harga_saham
})

df.head()
```

## Adding More Data

```python
harga_obligasi = np.random.randn(len(tanggal)).cumsum() + 70

df['harga_obligasi'] = harga_obligasi
df.head()
```

## Creating Subplot

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14,10))

axes[0].plot(df['tanggal'], df['harga_saham'], color='blue', label='Harga Saham')
axes[0].set_title('Harga Saham Selama Tahun 2022')
axes[0].set_ylabel('Harga Saham')
axes[0].grid(True)
axes[0].legend()

axes[1].plot(df['tanggal'], df['harga_obligasi'], color='red', label='Harga Obligasi')
axes[1].set_title('Harga Obligasi Selama Tahun 2022')
axes[1].set_xlabel('Tanggal')
axes[1].set_ylabel('Harga Obligasi')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```
