# Create Multiple Lineplot

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Dataframe

```python
tanggal = pd.date_range(start = '2021-01-01', end = '2024-01-01', freq = 'D')
harga_saham = np.random.randn(len(tanggal)).cumsum() + 100
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'tanggal': tanggal,
    'harga_saham': harga_saham
})

df.head()
```

## Adding More Data

```python
harga_obligasi = np.random.randn(len(tanggal)).cumsum() + 70

df['harga_obligasi'] = harga_obligasi
df.head()
```

## Different Lineplot With Rolling Mean

```python
df['30d_average'] = df['harga_saham'].rolling(window = 30).mean()

plt.figure(figsize=(14,7))
plt.plot(df['tanggal'], df['harga_saham'], color='blue', label='Harga Saham')
plt.plot(df['tanggal'], df['30d_average'], color='orange', linestyle='--', label = 'Rata-rata Bergerak 30 Hari')
plt.title('Harga Saham dengan Rata-rata Bergerak 30 Hari')
plt.xlabel('Tanggal')
plt.ylabel('Harga')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
