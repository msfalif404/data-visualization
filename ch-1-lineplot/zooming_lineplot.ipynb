# Create Multiple Lineplot

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Dataframe

```python
tanggal = pd.date_range(start = '2021-01-01', end = '2024-01-01', freq = 'D')
harga_saham = np.random.randn(len(tanggal)).cumsum() + 100
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'tanggal': tanggal,
    'harga_saham': harga_saham
})

df.head()
```

## Adding More Data

```python
harga_obligasi = np.random.randn(len(tanggal)).cumsum() + 70

df['harga_obligasi'] = harga_obligasi
df.head()
```

## Zooming Lineplot

```python
plt.figure(figsize=(14,7))
plt.plot(df['tanggal'], df['harga_saham'], color='blue', label='Harga Saham')
plt.plot(df['tanggal'], df['harga_obligasi'], color='red', linestyle='--', label='Harga Obligasi')
plt.title('Perbandinga Harga Saham dan Obligasi di Bulan Januari 2022')
plt.xlabel('Tanggal')
plt.ylabel('Harga')
plt.xlim(pd.Timestamp('2022-01-01'), pd.Timestamp('2022-01-31'))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
