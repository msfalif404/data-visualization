# Create Multiple Lineplot

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Dataframe

```python
bulan = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]
penjualan_produk_A = np.random.randint(50, 200, 12) * 100
penjualan_produk_B = np.random.randint(50, 200, 12) * 100
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'bulan': bulan,
    'produk_A': penjualan_produk_A,
    'produk_B': penjualan_produk_B
})

df.head()
```

## Compare Two Products

```python
plt.figure(figsize=(14,7))
plt.plot(df['bulan'], df['produk_A'], color='blue', marker='o', label='Produk A')
plt.plot(df['bulan'], df['produk_B'], color='red', marker='s', linestyle='--', label='Produk B')
plt.title('Perbandingan Penjualan Bulanan Produk A vs Produk B')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Showing Mean Of Two Products

```python
avg_A = df['produk_A'].mean()
avg_B = df['produk_B'].mean()

plt.figure(figsize=(14,7))
plt.plot(df['bulan'], df['produk_A'], color='blue', marker='o', label='Produk A')
plt.plot(df['bulan'], df['produk_B'], color='red', marker='s', linestyle='--', label='Produk B')

plt.axhline(avg_A, color='blue', linestyle='-.', label=f'Rata-rata Produk A: {avg_A:.2f}')
plt.axhline(avg_B, color='red', linestyle='-.', label=f'Rata-rata Produk B: {avg_B:.2f}')

plt.title('Perbandingan Rata-rata Penjualan Bulanan Produk A vs Produk B')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Visualization With Annotation

```python
bulan_peak_A = df['bulan'][df['produk_A'].idxmax()]
nilai_peak_A = df['produk_A'].max()

bulan_peak_B = df['bulan'][df['produk_B'].idxmax()]
nilai_peak_B = df['produk_B'].max()

plt.figure(figsize=(14,7))
plt.plot(df['bulan'], df['produk_A'], color='blue', marker='o', label='Produk A')
plt.plot(df['bulan'], df['produk_B'], color='red', marker='s', linestyle='--', label='Produk B')

plt.annotate(f'Peak Produk A: {bulan_peak_A}',
             xy=(bulan_peak_A, nilai_peak_A),
             xytext=(bulan_peak_A, nilai_peak_A + 2000),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=9)

plt.annotate(f'Peak Produk B: {bulan_peak_B}',
             xy=(bulan_peak_B, nilai_peak_B),
             xytext=(bulan_peak_B, nilai_peak_B + 2000),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=9)

plt.title('Perbandingan Penjualan Bulanan Produk A vs Produk B')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
