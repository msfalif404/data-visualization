# Creating Boxplot

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Data

```python
np.random.seed(42)

skor_murid_A = np.random.randint(60,90,30)
skor_murid_B = np.random.randint(55,95,30)
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'murid_A': skor_murid_A,
    'murid_B': skor_murid_B
})

df.head()
```

## Creating Boxplot

```python
plt.figure(figsize=(10,6))
plt.boxplot([df['murid_A'], df['murid_B']], vert=False, patch_artist=True, labels=['Murid A', 'Murid B'])
plt.title('Distribusi Skor Ujian Murid A dan Murid B')
plt.xlabel('Skor Ujian')
plt.ylabel('Kelompok Murid')
plt.grid(True, axis ='x', linestyle='--', linewidth=0.7, alpha=0.7)
plt.show()
```
