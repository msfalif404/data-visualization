# Basic Barchart

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Data

```python
genres = ['Fiksi','Non Fiksi','Komik','Pendidikan','Biografi','Horor','Romansa']
penjualan = [3450, 3120, 2980, 2760, 2530, 2350, 2100]
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'genre': genres,
    'penjualan': penjualan
})

df.head()
```

## Creating Barchart

```python
plt.figure(figsize=(10,6))
plt.bar(df['genre'], df['penjualan'], color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Penjualan Buku Berdasarkan Genre Selama Setahun Terakhir')
plt.xlabel('Genre Buku')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, axis = 'y')
plt.tight_layout()
plt.show()
```
