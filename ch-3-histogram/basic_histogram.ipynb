# Basic Histogram

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthethic Data

```python
np.random.seed(42)
umur_pelanggan = np.random.randint(15,75,1000)
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'umur': umur_pelanggan
})

df.head()
```

## Creating Histogram

```python
plt.figure(figsize=(10,6))
plt.hist(df['umur'], bins=12, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribusi Umur Pelanggan')
plt.xlabel('Umur')
plt.ylabel('Jumlah Pelanggan')
plt.grid(True, which='both', linestyle='--', linewidth='0.5')
plt.tight_layout()
plt.show()
```
