# Basic Scatterplot

## Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating Synthetic Data

```python
np.random.seed(42)
usia = np.random.randint(20,60,100)
penghasilan = (usia * 500) + np.random.randn(100) * 3000
```

## Creating Dataframe

```python
df = pd.DataFrame({
    'usia': usia,
    'penghasilan': penghasilan
})

df.head()
```

## Creating Scatter Plot

```python
plt.figure(figsize=(10,6))
plt.scatter(df['usia'], df['penghasilan'], color='purple', alpha=0.6, edgecolors='white')
plt.title('Hubungan Antara Usia dan Penghasilan')
plt.xlabel('Usia')
plt.ylabel('Penghasilan')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
